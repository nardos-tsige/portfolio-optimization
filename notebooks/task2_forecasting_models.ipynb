```python
!pip install pmdarima statsmodels tensorflow scikit-learn yfinance -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime, timedelta
import warnings
import os
import zipfile
warnings.filterwarnings('ignore')

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
os.makedirs('data/processed/', exist_ok=True)

print("=" * 60)
print("TASK 2: TIME SERIES FORECASTING MODELS")
print("=" * 60)

print("\n1. DOWNLOADING DATA...")
start_date = '2015-01-01'
end_date = '2026-06-30'

print("Fetching TSLA...")
tsla_raw = yf.download('TSLA', start=start_date, end=end_date, progress=False)
print("Fetching BND...")
bnd_raw = yf.download('BND', start=start_date, end=end_date, progress=False)
print("Fetching SPY...")
spy_raw = yf.download('SPY', start=start_date, end=end_date, progress=False)

print("\n2. EXTRACTING CLOSE PRICES AS NUMERIC SERIES...")
tsla_close = tsla_raw['Close'].values.flatten()
bnd_close = bnd_raw['Close'].values.flatten()
spy_close = spy_raw['Close'].values.flatten()

print(f"TSLA close shape: {tsla_close.shape}")

tsla_dates = tsla_raw.index
tsla_series = pd.Series(tsla_close, index=tsla_dates, dtype=float)
print(f"TSLA series type: {tsla_series.dtype}")

print("\n3. SPLITTING DATA...")
train_size = int(len(tsla_series) * 0.8)
train_data = tsla_series[:train_size]
test_data = tsla_series[train_size:]

print(f"Training: {len(train_data)} samples")
print(f"Testing: {len(test_data)} samples")

print("\n4. STATIONARITY TEST...")
def adf_test(series, title="Series"):
    if isinstance(series, pd.Series):
        series = series.values
    series_clean = series[~np.isnan(series)]
    print(f"{title} - Length: {len(series_clean)}")
    if len(series_clean) < 10:
        print("Not enough data for ADF test")
        return False
    result = adfuller(series_clean, autolag='AIC')
    print(f'ADF Statistic: {result[0]:.6f}')
    print(f'p-value: {result[1]:.6f}')
    print('Critical values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value:.3f}')
    return result[1] <= 0.05

train_values = train_data.values
is_stationary = adf_test(train_values, "Training data")
print(f"Is stationary: {is_stationary}")

if not is_stationary:
    print("\nDifferencing data to make it stationary...")
    train_diff = np.diff(train_values)
    print("Testing differenced data:")
    adf_test(train_diff, "Differenced data")

print("\n5. FINDING OPTIMAL ARIMA PARAMETERS...")
print("This may take a few minutes...")

train_clean = train_values[~np.isnan(train_values)]

auto_model = auto_arima(
    train_clean,
    start_p=0, start_q=0,
    max_p=5, max_q=5,
    seasonal=False,
    trace=False,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True,
    random_state=42
)

p, d, q = auto_model.order[0], auto_model.order[1], auto_model.order[2]
print(f"Best ARIMA parameters: (p={p}, d={d}, q={q})")
print(f"Best AIC: {auto_model.aic():.4f}")

print("\n6. FITTING ARIMA MODEL...")
arima_model = ARIMA(train_clean, order=(p, d, q))
arima_fit = arima_model.fit()
print(arima_fit.summary())

print("\n7. ARIMA FORECAST...")
test_values = test_data.values
forecast_steps = len(test_values)
arima_forecast = arima_fit.forecast(steps=forecast_steps)
arima_forecast_index = test_data.index

test_clean = test_values[~np.isnan(test_values)]
arima_mae = mean_absolute_error(test_clean, arima_forecast)
arima_rmse = np.sqrt(mean_squared_error(test_clean, arima_forecast))
arima_mape = np.mean(np.abs((test_clean - arima_forecast) / test_clean)) * 100

print(f"ARIMA - MAE: ${arima_mae:.2f}, RMSE: ${arima_rmse:.2f}, MAPE: {arima_mape:.2f}%")

print("\n8. PLOTTING ARIMA...")
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(train_data.index, train_values, label='Training Data', linewidth=1.5)
ax.plot(test_data.index, test_values, label='Actual Test Data', linewidth=1.5)
ax.plot(arima_forecast_index, arima_forecast, label='ARIMA Forecast', linewidth=2, linestyle='--')
ax.set_title('ARIMA Model: Tesla Close Price Forecast', fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Price ($)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('data/processed/arima_forecast.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: data/processed/arima_forecast.png")

print("\n9. PREPARING LSTM DATA...")
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(tsla_series.values.reshape(-1, 1))

def create_sequences(data, window_size=60):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

window_size = 60
X, y = create_sequences(scaled_data, window_size)

train_size_lstm = int(len(X) * 0.8)
X_train, X_test = X[:train_size_lstm], X[train_size_lstm:]
y_train, y_test = y[:train_size_lstm], y[train_size_lstm:]

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

print(f"Training shape: {X_train.shape}")
print(f"Testing shape: {X_test.shape}")

print("\n10. BUILDING LSTM MODEL...")
def build_lstm_model(window_size):
    model = Sequential([
        LSTM(units=50, return_sequences=True, input_shape=(window_size, 1)),
        Dropout(0.2),
        LSTM(units=50, return_sequences=False),
        Dropout(0.2),
        Dense(units=25),
        Dense(units=1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

lstm_model = build_lstm_model(window_size)
lstm_model.summary()

print("\n11. TRAINING LSTM...")
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = lstm_model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=1
)

print("\n12. PLOTTING LSTM TRAINING HISTORY...")
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(history.history['loss'], label='Training Loss')
ax.plot(history.history['val_loss'], label='Validation Loss')
ax.set_title('LSTM Training History', fontsize=14, fontweight='bold')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('data/processed/lstm_training_history.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: data/processed/lstm_training_history.png")

print("\n13. LSTM PREDICTIONS...")
lstm_pred_scaled = lstm_model.predict(X_test)
lstm_pred = scaler.inverse_transform(lstm_pred_scaled)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Get correct dates for LSTM predictions
# The test data starts after train_size + window_size
lstm_test_start = train_size_lstm + window_size
lstm_test_end = lstm_test_start + len(lstm_pred)
lstm_test_dates = tsla_series.index[lstm_test_start:lstm_test_end]

print(f"LSTM predictions shape: {lstm_pred.shape}")
print(f"LSTM dates shape: {len(lstm_test_dates)}")

# Flatten predictions
lstm_pred = lstm_pred.flatten()
y_test_actual = y_test_actual.flatten()

print(f"After flatten - predictions: {lstm_pred.shape}, actual: {y_test_actual.shape}")

# Trim to match lengths
min_len = min(len(lstm_pred), len(y_test_actual))
lstm_pred = lstm_pred[:min_len]
y_test_actual = y_test_actual[:min_len]
lstm_test_dates = lstm_test_dates[:min_len]

lstm_mae = mean_absolute_error(y_test_actual, lstm_pred)
lstm_rmse = np.sqrt(mean_squared_error(y_test_actual, lstm_pred))
lstm_mape = np.mean(np.abs((y_test_actual - lstm_pred) / y_test_actual)) * 100

print(f"LSTM - MAE: ${lstm_mae:.2f}, RMSE: ${lstm_rmse:.2f}, MAPE: {lstm_mape:.2f}%")

print("\n14. PLOTTING LSTM...")
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(train_data.index, train_values, label='Training Data', linewidth=1.5)
ax.plot(test_data.index, test_values, label='Actual Test Data', linewidth=1.5)
ax.plot(lstm_test_dates, lstm_pred, label='LSTM Prediction', linewidth=2, linestyle='--')
ax.set_title('LSTM Model: Tesla Close Price Prediction', fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Price ($)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('data/processed/lstm_prediction.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: data/processed/lstm_prediction.png")

print("\n15. MODEL COMPARISON...")
comparison_data = {
    'Model': ['ARIMA', 'LSTM'],
    'MAE ($)': [arima_mae, lstm_mae],
    'RMSE ($)': [arima_rmse, lstm_rmse],
    'MAPE (%)': [arima_mape, lstm_mape]
}
comparison_df = pd.DataFrame(comparison_data)
print("=" * 60)
print(comparison_df.to_string(index=False))
print("=" * 60)

best_model = 'LSTM' if lstm_mae < arima_mae else 'ARIMA'
print(f"\nBest performing model: {best_model}")
comparison_df.to_csv('data/processed/model_comparison.csv', index=False)

print("\n16. COMBINED PLOT...")
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(train_data.index, train_values, label='Training Data', linewidth=1.5, alpha=0.7)
ax.plot(test_data.index, test_values, label='Actual Test Data', linewidth=1.5)
ax.plot(arima_forecast_index, arima_forecast, label='ARIMA Forecast', linewidth=2, linestyle='--')
ax.plot(lstm_test_dates, lstm_pred, label='LSTM Prediction', linewidth=2, linestyle='-.')
ax.set_title('Model Comparison: Tesla Close Price Predictions', fontsize=14, fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Price ($)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('data/processed/model_comparison_plot.png', dpi=300, bbox_inches='tight')
plt.show()
print("Saved: data/processed/model_comparison_plot.png")

print("\n17. SAVING AND DOWNLOADING RESULTS...")
with zipfile.ZipFile('task2_results.zip', 'w') as zipf:
    for file in os.listdir('data/processed/'):
        if file.endswith('.png') or file.endswith('.csv'):
            zipf.write(f'data/processed/{file}', file)

from google.colab import files
files.download('task2_results.zip')
print("Downloaded: task2_results.zip")

print("\n" + "=" * 60)
print("TASK 2 COMPLETE!")
print("=" * 60)
print(f"Best performing model: {best_model}")
print(f"ARIMA MAPE: {arima_mape:.2f}%")
print(f"LSTM MAPE: {lstm_mape:.2f}%")
```
